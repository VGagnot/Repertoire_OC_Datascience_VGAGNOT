```python
def group_cat(row):
    y = 'Autres'

    if row['product_category_name_english'] in [
                        'musical_instruments',
                        'cine_photo',
                        'music',
                        'books_technical',
                        'books_general_interest',
                        'books_imported,dvds_blu_ray',
                        'cds_dvds_musicals',
                        'arts_and_craftmanship',
                        'sports_leisure',
                        'art'
                        ]:
        y = 'Culture&arts'

    if row['product_category_name_english'] in [
                        'furniture_decor',
                        'garden_tools',
                        'bed_bath_table',
                        'home_appliances',
                        'kitchen_dining_laundry_garden_furniture',
                        'home_confort',
                        'costruction_tools_garden',
                        'construction_tools_construction',
                        'construction_tools_lights',
                        'home_construction,construction_tools_safety',
                        'furniture_living_room',
                        'furniture_bedroom',
                        'home_comfort_2',
                        'furniture_mattress_and_upholstery',
                        'costruction_tools_tools',
                        'housewares',
                        'flowers'
                        ]:
        y = 'Deco&maison'

    if row['product_category_name_english'] in [
                        'telephony',
                        'consoles_games',
                        'electronics',
                        'computers_accessories',
                        'audio',
                        'fixed_telephony',
                        'air_conditioning',
                        'small_appliances_home_oven_and_coffee',
                        'signaling_and_security',
                        'small_appliances',
                        'home_appliances_2',
                        'computers',
                        'tablets_printing_image'
                        ]:
        y = 'Electronique'

    if row['product_category_name_english'] in [
                        'fashion_bags_accessories',
                        'luggage_accessories',
                        'fashion_underwear_beach',
                        'fashion_male_clothing',
                        'fashion_shoes',
                        'fashio_female_clothing',
                        'fashion_sport',
                        'fashion_childrens_clothes'
                        ]:
        y = 'Accessoires&vetements'

    if row['product_category_name_english'] in [
                        'perfumery',
                        'health_beauty',
                        'diapers_and_hygiene'
                        ]:
        y = 'Hygiene&cosmetiques'

    if row['product_category_name_english'] in [
                        'food',
                        'food_drink',
                        'drinks',
                        'la_cuisine'
                        ]:
        y = 'Alimentaire'

    if row['product_category_name_english'] in [
                        'cool_stuff',
                        'pet_shop',
                        'office_furniture',
                        'agro_industry_and_commerce',
                        'baby',
                        'toys',
                        'stationery',
                        'industry_commerce_and_business',
                        'auto, party_supplies',
                        'market_place',
                        'christmas_supplies',
                        'security_and_services',
                        'watches_gifts',
                        'auto',
                        'la_cuisine'
                        ]:
        y = 'Autres'
    return y
```

```python
data0_order_items['product_cat'] = data0_order_items.apply(
                                        lambda x: group_cat(x), axis=1)
```

```python
def to_timestamp(row):
    y = 0
    x = row['order_purchase_timestamp']
    y = datetime.datetime.strptime(x[0:10], "%Y-%m-%d")
    z = (y - datetime.datetime.now()).days
    return z
```

## K-means

### Recherche des meilleurs paramètres

```python
n_clusters = range(2, 15)
score_silhouette_n = []
inertie_n = []

for n in n_clusters:

    kmeans = KMeans(
        init="random",
        n_clusters=n,
        n_init=10,
        max_iter=300,
        random_state=42
    )

    kmeans.fit(data1_orders_by_customer_unique_id_normalized[[
        'nb_orders',
        'date_timestamp',
        'payment_value',
        'review_score'
    ]])

    clusters = kmeans.predict(data1_orders_by_customer_unique_id_normalized[[
        'nb_orders',
        'date_timestamp',
        'payment_value',
        'review_score'
    ]])

    data1_orders_by_customer_unique_id['cluster'] = clusters

    x = silhouette_score(data1_orders_by_customer_unique_id_normalized[[
        'nb_orders',
        'date_timestamp',
        'payment_value',
        'review_score'
    ]],
        data1_orders_by_customer_unique_id['cluster']
    )

    score_silhouette_n.append(x)

    y = kmeans.inertia_
    inertie_n.append(y)
```

```python
ax = plt.gca()

ax.plot(n_clusters, score_silhouette_n)
plt.xlabel('n_clusters')
plt.ylabel('score_silhouette')
plt.title('xxx')
plt.show()
```

```python
ax = plt.gca()

ax.plot(n_clusters, inertie_n)
plt.xlabel('n_clusters')
plt.ylabel('inertie_n')
plt.title('xxx')
plt.show()
```

Les résultats pour plus de 3 clusters sont meilleurs que l'Agglomerative Clustering, et toujours meilleurs que le DBScan.

## Un indicateur de similarité: Matrice de similarité

```python
def mat_sim(a, b):

    # Cette fonction compare 2 clusterings.
    # Elle fonctionne sur un principe simple: dans une matrice de confusion,
    # le maximum d'occurences signifie que les 2 clusters associés sont
    # en fait le même cluster.
    # On peut alors éliminer ces 2 clusters de l'analyse et de la matrice
    # de confusion, puis recommencer l'identification.
    # La fonction renvoie:
    # 1) un score de similarité entre 0 (complètement différent) et 1
    #  (absolument identique)
    # 2) la correspondance des clusters entre les 2 clusterings

    import numpy as np
    from sklearn.metrics import confusion_matrix

    # On vérifie que les 2 listes ont le même nombre de clusters
    if len(np.unique(a)) != len(np.unique(b)):
        x = 'nombres de clusters différents!'
        y = 'nombres de clusters différents!'
    else:

        # Si tout va bien, on initie - et on calcule la matrice de confusion.
        # La copie initiale sera conservée (mat1), une 2e copie sera modifiée
        # au fur et à mesure (mat2).
        # La liste 'correspondance' identifiera les correspondances de clusters
        # entre les 2 entrées, sur la base de l'index dans la matrice
        # L'entier 'ok' sommera le nombre d'occurences pour les
        # correspondances sélectionnées.
        correspondance = []
        mat1 = confusion_matrix(a, b)
        mat2 = mat1
        ok = 0
        for i in range(0, len(np.unique(a)) - 1):
            mx = np.amax(mat2)
            if len(np.where(mat2 == mx)[0]) > 1:
                for j in (0, len(np.where(mat2 == mx)[0]) - 1):
                    if any(mx in l for l in mat2):
                        x1 = np.where(mat2 == mx)[0][0]
                        x2 = np.where(mat2 == mx)[1][0]
                        correspondance.append(
                            [np.where(mat1 == mx)[0][j],
                             np.where(mat1 == mx)[1][j]]
                        )
                        ok = ok + mx
                        mat2 = np.delete(mat2, x1, axis=0)
                        mat2 = np.delete(mat2, x2, axis=1)
            else:
                x1 = np.where(mat2 == mx)[0][0]
                x2 = np.where(mat2 == mx)[1][0]
                correspondance.append([
                    np.where(mat1 == mx)[0][0],
                    np.where(mat1 == mx)[1][0]])
                ok = ok + mx
                mat2 = np.delete(mat2, x1, axis=0)
                mat2 = np.delete(mat2, x2, axis=1)
        x = ok/np.sum(mat1)
        y = pd.DataFrame(
                correspondance,
                columns=['cluster1', 'cluster2'])
    return[x, y]
```

Je mets en place une fonction pour, à partir de la base de données des commandes à aujourd'hui, pouvoir simuler la base des commandes il y a j jours.  
Je pourrai ainsi utiliser cette fonction dans une boucle.  
  
La fonction aura, comme variable de sortie:  
0) la base de données client il y a j jours, y compris le clustering effectué à cette date  
1) le kmeans fitté à cette date, pour 4 variables ('nb_orders', 'payment_value','review_score', 'date_timestamp')  
2) le score_silhouette correspondant à 1) et 2)  
3) la normalisation fittée, pour les 4 variables mentionnées en 1) et en date -j jours.

```python
derniere_date = max(data2_orders['date_timestamp'])
```

```python
def db_j_jours(db, j):
    # A partir d'un dataframe db, cette fonction remonte le temps de j jours
    # (supprime les j derniers jours), et renvoie:
    # 0) la base de données client correspondante, y compris le clustering
    # 1) le kmeans fitté, pour 4 variables ('nb_orders', 'payment_value',
    # 'review_score', 'date_timestamp')
    # 2) le score_silhouette correspondant à 1) et 2)
    # 3) la normalisation fittée, pour les 4 variables mentionnées en 1)

    # Déroulement:
    # On retire les x derniers jours de la base des orders
    data_orders_j = db.drop(
        db[db['date_timestamp'] >= derniere_date + j].index)

    # On regroupe les données par client
    data_orders_j_by_cust = data_orders_j[[
        'customer_unique_id',
        'nb_orders',
        'payment_value',
        'review_score',
        'date_timestamp'
    ]].groupby('customer_unique_id').agg({
        'date_timestamp': 'max',
        'payment_value': 'sum',
        'review_score': 'mean',
                        'nb_orders': 'sum'
    })

    # On normalise la bdd obtenue

    std_scl_j = std_scaler.fit(data_orders_j_by_cust[[
        'nb_orders',
        'payment_value',
        'review_score',
        'date_timestamp'
    ]]
    )

    data_scaled = std_scl_j.transform(data_orders_j_by_cust[[
        'nb_orders',
        'payment_value',
        'review_score',
        'date_timestamp'
    ]]
    )

    data_orders_j_by_cust_normalized = pd.DataFrame(
        data_scaled,
        columns=[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]]
    )

    # On paramètre une nouvelle version du KMeans, puis on la fitte
    kmeans_j = KMeans(
        init="random",
        n_clusters=n,
        n_init=10,
        max_iter=300,
        random_state=42
    )

    kmeans_j.fit(
        data_orders_j_by_cust_normalized[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]]
    )

    # On calcule les clusters de chaque client, à partir de la bdd "passé"
    # et du KMeans
    clusters = kmeans_j.predict(
        data_orders_j_by_cust_normalized[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]]
    )

    data_orders_j_by_cust['cluster_j'] = clusters

    # On calcule le score silhouette correspondant
    silhouette_j = silhouette_score(data_orders_j_by_cust_normalized[[
        'nb_orders',
        'payment_value',
        'review_score',
        'date_timestamp'
    ]],
        clusters
    )

    # On choisit les sorties
    return [data_orders_j_by_cust, kmeans_j, silhouette_j, std_scl_j]
```

Je choisis les différentes dates auxquelles je veux simuler le clustering (retour j jours en arrière):

```python
retour = [50, 55, 60, 65, 66, 67, 68, 69, 70,
          71, 72, 73, 74, 75, 80, 85, 90, 95, 100]
```

```python
jours = []
s_score1 = []
s_score2 = []
ARI1 = []
ARI2 = []
similarity1 = []
similarity2 = []


for j in retour:

    # Création et utilisation de la base passé simulée
    bloup = db_j_jours(data2_orders, -j)

    # Liste des nbres de jour
    jours.append(j)

    # avec nouvelle normalisation et ancien clustering: silhouette score et ARI
    clusters = bloup[1].predict(
        data1_orders_by_customer_unique_id_normalized[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]]
    )

    data1_orders_by_customer_unique_id['cluster1'] = clusters

    bloup[0] = bloup[0].merge(
        data1_orders_by_customer_unique_id[['cluster1']],
        left_index=True,
        right_index=True
    )

    s_score1.append(
        silhouette_score(data1_orders_by_customer_unique_id_normalized[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]],
            clusters
        ))

    ARI1.append(
        adjusted_rand_score(
            bloup[0]['cluster_j'],
            bloup[0]['cluster1']
        ))

    similarity1.append(mat_sim(
        bloup[0]['cluster_j'],
        bloup[0]['cluster1']
    )[0])

    # avec ancienne normalisation et ancien clustering: silhouette score ,
    # ARI, score de similarité

    data_scaled = bloup[3].transform(data1_orders_by_customer_unique_id[[
        'nb_orders',
        'payment_value',
        'review_score',
        'date_timestamp'
    ]]
    )

    db_data_scaled = pd.DataFrame(
        data_scaled,
        columns=[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]]
    )

    clusters = bloup[1].predict(
        db_data_scaled[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]]
    )

    data1_orders_by_customer_unique_id['cluster2'] = clusters

    bloup[0] = bloup[0].merge(
        data1_orders_by_customer_unique_id[['cluster2']],
        left_index=True,
        right_index=True
    )

    s_score2.append(
        silhouette_score(data1_orders_by_customer_unique_id_normalized[[
            'nb_orders',
            'payment_value',
            'review_score',
            'date_timestamp'
        ]],
            clusters
        ))

    ARI2.append(adjusted_rand_score(
        bloup[0]['cluster_j'],
        bloup[0]['cluster2']
    ))

    similarity2.append(mat_sim(
        bloup[0]['cluster_j'],
        bloup[0]['cluster2']
    )[0])

    # On supprime les 2 nouvelles colonnes de clusters ajoutées dans ce cycle

    data1_orders_by_customer_unique_id.drop(columns=['cluster1', 'cluster2'])
```
